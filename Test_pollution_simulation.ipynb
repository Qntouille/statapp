```python
import numpy as np
import numpy.random as rd

# Ici je génère aléatoirement une matrice 20 par 20 qui a les mêmes propriétés que la matrice flux rouges
# c'est à dire des coeffs diagonaux assez élevés et le reste pour la plupart égaux à 0
matrix=np.zeros((20,20))
for i in range(20):
    for j in range(20):
        if i==j : 
            matrix[i, j] = rd.randint(200, 1500)
        else : 
            matrix[i, j] = rd.randint(0, 150)
print(matrix)

# Ici je fais un numpy array qui contient les populations des villes (il y a 45,6% d'actifs en France
# donc je fais l'inverse de ça fois le nombre de personnes qui se déplacent au départ d'une ville)
population=np.zeros(20)
for i in range(20):
    nbactifs=0
    for j in range(20):
        nbactifs+=matrix[i,j]
    population[i]=(1000/456)*nbactifs

# Là je génère des superficies aléatoires pour mes villes, ce sera utile pour calculer la distance moyenne
# entre deux points aléatoires au sein d'une même ville quand on voudra prendre en compte la pollution engendrée 
# par les gens qui se déplacent avec leur voiture au sein de leur ville pour aller travailler
city_areas = np.zeros(20)
for i in range(20):
    city_areas[i] = rd.randint(10, 1000)

# Là je définis des distances entre les villes, elles sont aléatoires et je fixe à 0 la distance entre deux villes 
# pour lesquelles personne ne fait le déplacement, pour simplifier
dist=np.zeros((20,20))
for i in range(20):
    for j in range(20):
        if matrix[i,j]==0 :
            dist[i,j]=0
        else : dist[i,j]=rd.randint(3,50)

CO2km=0.11
# Une voiture en gros ça émet 110g de CO2 par km parcouru

# Là je fais un tableau qui contient la pollution de chaque ville, en tant que ville et deépart ET d'arrivée
A=np.zeros(20)
for i in range(20):
    for j in range(20): 
        if i != j :
            A[i]+=dist[i,j]*CO2km*matrix[i,j] #on prend toutes les émissions dans lesquelles chaque ville est impliquée, on compte donc tout en double
            # distance fois nb de personnes qui font le déplacement fois émission au km
        else : 
            A[i]+=0.9054*CO2km*np.sqrt(city_areas[i]/np.pi)
            # la distance moyenne entre deux points aléatoires dans un cercle (lieu de départ et lieu d'arrivée
            # au sein d'une même ville) est 0.9054 fois le rayon. J'assimile les villes à des cercles pour simplifier
print(A)

pollution_par_personne=[A[i]/population[i] for i in range(20)]
# Là je rapporte la pollution engendrée à la population de la ville pour ne pas avoir des trucs aberrants du genre
# Paris = la plus peuplée donc logiquement pollue plus que la plupart des autres villes même si je dis n'importe
# quoi elle fait partie de la moitié des villes où les gens individuellement polluent le moins

print(f"La ville qui pollue le plus rapporté à sa population est la ville n°", np.argmax(pollution_par_personne)+1)
# Vu que tout est généré aléatoirement le résultat ne veut rien dire mais dans l'idée est-ce que ce genre de 
# raisonnement ça vous irait pour l'appliquer à notre tableau flux rouges ?
```
